```python
from bs4 import BeautifulSoup as bs
import requests as rq
from urllib.parse import urlparse
import os
```

```python
#网页获取、编码以及源代码
url = 'https://www.fabiaoqing.com/'
html = rq.get(url)
html.encoding = 'utf-8'
print(html.text)
```

```python
sp = bs(html.text,'html.parser')
print(sp)
```

```python
emoji_img = sp.select('img')
print(emoji_img)
```

```python
#获取数据
img_list = []
for img in sp.select('img'):
    #print(img)
    #print(img.has_attr('src'))
    print(img.attrs['src'])
    if img.has_attr("src"):
        img_list.append(img.attrs['src'])
    else:
        img_list.append(img.attrs['src'])
```

```python
img_dir = os.path.join(os.curdir, "img")
if not os.path.isdir(img_dir):
    os.mkdir(img_dir)
```

```python
o = urlparse(img_list[2])
print(o.path)
print(type(o.path))
```

```python
# 下载图片
i = 0
for img in img_list:
    print(i+1)
    if img == '':
        i+=1
        continue
    if img[0] == '/':
        i+=1
        continue
    #print(type(img))
    o = urlparse(img)
    url = '%s://%s/%s' % (o.scheme, o.netloc, o.path)
    #print(url)
    if True:
        
        # 图片的二进制数据
        resp = rq.get(url)
        #print('1')
        img_name = str(i)
        img_path = os.path.join(img_dir, img_name)
        # 把二进制数据写入图片中
        #print(img_path)
        with open(img_path, "wb") as f:
            # 每次写1024个字节
            for c in resp.iter_content(1024):
                f.write(c)
    i += 1

print('end')
```
